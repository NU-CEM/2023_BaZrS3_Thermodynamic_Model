```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
BZS = []
BaS = []
ZrS2 =[]
Ba =[]
Zr =[]
S = []
temp=[]

with open('./data/phonopy_output/BaZrS3_Pnma.dat', 'r') as mode1:
    for line in mode1:
        line=line.split()
        if line[0] != '#':
            BZS.append(float(line[2]))
with open('./data/phonopy_output/BaS_Fm-3m.dat', 'r') as mode1:
    for line in mode1:
        line=line.split()
        if line[0] != '#':
            BaS.append(float(line[2]))
with open('./data/phonopy_output/ZrS2_P-3m1.dat', 'r') as mode1:
    for line in mode1:
        line=line.split()
        if line[0] != '#':
            ZrS2.append(float(line[2]))
with open('./data/phonopy_output/Ba_Im-3m.dat', 'r') as mode1:
    for line in mode1:
        line=line.split()
        if line[0] != '#':
            Ba.append(float(line[2]))
with open('./data/phonopy_output/S_P2_c.dat', 'r') as mode1:
    for line in mode1:
        line=line.split()
        if line[0] != '#':
            S.append(float(line[2]))
with open('./data/phonopy_output/Zr_P63_mmc.dat', 'r') as mode1:
    for line in mode1:
        line=line.split()
        if line[0] != '#':
            temp.append(float(line[0]))
            Zr.append(float(line[2]))

```

```python
plt.plot(temp,np.array(Ba)/1,label='Ba')  # divided by number of atoms in cell.
plt.plot(temp,np.array(Zr)/2,label='Zr')
plt.plot(temp,np.array(S)/32,label='S')
plt.plot(temp, np.array(BZS)/20,label='BZS')
plt.title("Entropy of perovskite and solid elemental precursors")
plt.ylabel('$\Delta$ S (kJ/mol)')
plt.xlabel('Temperature (K)')
plt.legend()

# makes sense that entropic contributionsare highest to lowest for the heaviest to lightest atoms, with BZS as a 
```

```python
plt.plot(temp,np.array(BaS)/2,label='BaS')  # divided by number of atoms in cell.
plt.plot(temp,np.array(ZrS2)/3,label='ZrS2')
plt.plot(temp, np.array(BZS)/20,label='BZS')
plt.title("Entropy of perovskite and binary precursors")
plt.ylabel('$\Delta$ S (kJ/mol)')
plt.xlabel('Temperature (K)')
plt.legend()

# also makes sense looking at average mass of atoms.
```

```python
change_S_binary_sulfides = (np.array(BZS)/4) - np.array(BaS) - np.array(ZrS2)  # factor of 4 to mass balance
change_S_elemental = (np.array(BZS)/4) - (np.array(Ba)) - (np.array(Zr)/2) - (np.array(S)*(3/32))

plt.plot(temp,change_S_binary_sulfides, label='perovskite formation from binary sulphides')
plt.plot(temp,change_S_elemental, label='perovskite formation from constituent elements')
plt.legend()
plt.title("Change in entropy when forming perovskite")
plt.ylabel('$\Delta$ S (kJ/mol)')
plt.xlabel('Temperature (K)')


# this shows that the change in entropy (Delta S) is positive for formation from binary sulfides. 
# there is an entropy gain: displacements from equilibrium tend to be larger.
# ---> perovskite becomes increasingly stabilised with temperature.


# in contrast, the change in entropy (Delta S) is negative for formation from solid elemental compounds.
# there is an entropy loss: displacements from equilbrium are reduced.
# ---> as temperature increases the perovskite becomes less stable.
```
